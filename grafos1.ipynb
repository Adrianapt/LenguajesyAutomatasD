# Instalación y pruebas

```python
import networkx as nx
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
#aqui definimos el grafo vacio
G = nx.Graph()
```

```python
#sirve para ver la cant de vertices
G.order()
```

```python
#los lados que tiene o tamaÑo
G.size()
```

```python
#es para crear una lista
vertices_G = ['a','b','c']
G.add_nodes_from(vertices_G)
```

```python
G.order()
```

```python
lados_G = [('a', 'b'), ('a','c')]
G.add_edges_from(lados_G)
```

```python
G.size()
```

```python
#solo los lados de a en este caso
nx.degree(G,'a')
```

```python
#aqui muestra en forma de lista los grados 
list(G.degree())
```

```python
#con esto agregamos
G.add_edge('c','b')
```

```python
list (G.degree())
```

## Forma Grafica

```python
nx.draw (G, with_labels= True,
        node_color = 'yellow')
```

```python
nx.draw_spring (G,node_color = 'cyan', node_size = 5000)
```

```python
#ubicacione pra cambiar coordenadas de mi grafo
ubica = {'a': (0,0), 'b': (1,1), 'c':(-2,1)}
nx.draw (G, pos=ubica, with_labels = True, node_color = 'yellow')
```

```python
#quitar un lado
G.remove_edge('c','a')
nx.draw (G, pos=ubica, with_labels = True, node_color = 'red')
```

```python
G.remove_node('c')
nx.draw (G, pos=ubica, with_labels = True, node_color = 'yellow')
```

```python
#grafo bipartite
B = nx.complete_bipartite_graph(4,6)
nx.draw (B ,with_labels = True, node_color = 'red')
```

```python
X, Y = nx.bipartite.sets(B)
```

```python
ubica_B = nx.bipartite_layout (B,list(X), align = 'horizontal')
nx.draw (B , pos=ubica_B,with_labels = True, node_color= 'yellow')
```
